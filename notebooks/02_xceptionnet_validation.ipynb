# 02 — XceptionNet validation (Phase 3)

Pretrained **FaceForensics++** Xception (`full_c23.p`): **299×299** input, **mean/std = 0.5** (not ImageNet).

1. Download weights per `docs/PROJECT_PLAN_v10.md` §6 (`faceforensics++_models.zip` → `models/…/full_c23.p`).
2. Run cells below; for full benchmark use `training/evaluate_spatial_xception.py` on GPU after face extraction.

```python
from pathlib import Path
import numpy as np

from src.modules.spatial import SpatialDetector

def find_full_c23(models_dir: Path = Path("../models")) -> Path | None:
    for p in models_dir.resolve().rglob("full_c23.p"):
        return p
    return None

w = find_full_c23()
if w is None:
    print("Place full_c23.p under models/ (see PROJECT_PLAN_v10 §6).")
else:
    det = SpatialDetector(w, device="cpu")
    crop = np.random.randint(0, 255, (299, 299, 3), dtype=np.uint8)
    p = det.predict_frame(crop)
    assert 0.0 <= p <= 1.0
    print("weights:", w)
    print("P(fake) sample:", p)
```
