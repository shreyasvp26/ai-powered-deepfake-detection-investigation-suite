# 03 — Temporal analysis (Phase 3)

Four features over per-frame Xception **P(fake)** (see `docs/PROJECT_PLAN_v10.md` §7): `global_variance`, `sign_flip_rate`, `max_window_variance`, `max_jump`.

```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt

from src.modules.temporal import TemporalAnalyzer

cfg = Path("../configs/inference_config.yaml")
ta = TemporalAnalyzer(inference_config_path=cfg if cfg.exists() else None)

stable = [0.42] * 25
unstable = [0.1 if i % 2 == 0 else 0.9 for i in range(25)]

for name, series in [("stable", stable), ("unstable", unstable)]:
    r = ta.analyze(series)
    print(name, "Ts=", r["temporal_score"], "sign_flip=", r["sign_flip_rate"])

fig, ax = plt.subplots(figsize=(8, 3))
ax.plot(unstable, marker=".")
ax.set_ylabel("P(fake)")
ax.set_title("Synthetic oscillating per-frame scores")
plt.tight_layout()
plt.show()
```
