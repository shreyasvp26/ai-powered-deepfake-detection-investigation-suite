# 01 — Data exploration (FF++ splits + crop tree sanity)\n\nThis notebook is designed to run **without GPU**. It validates the FF++ split JSONs and (if available) the extracted face crop directory layout.

```python
from pathlib import Path
import json

SPLITS = Path('data/splits')
FACES_ROOT = Path('data/processed/faces')

def load_pairs(path: Path):
    with path.open() as f:
        pairs = json.load(f)
    pairs = [p for p in pairs if isinstance(p, list) and len(p)==2]
    return pairs

for name in ['train.json','val.json','test.json','train_identity_safe.json','val_identity_safe.json','test_identity_safe.json']:
    p = SPLITS / name
    if p.exists():
        pairs = load_pairs(p)
        print(f"{name}: {len(pairs)} pairs")
    else:
        print(f"{name}: MISSING")
```

```python
# Real IDs split (needed for fusion feature extraction)
real_ids_path = SPLITS / 'real_source_ids_identity_safe.json'
if real_ids_path.exists():
    real_ids = json.loads(real_ids_path.read_text())
    for k,v in real_ids.items():
        print(k, len(v), 'examples:', v[:5])
else:
    print('Missing:', real_ids_path)
```

```python
# Crop tree sanity checks (optional if you don't have crops locally)
methods = ['original', 'Deepfakes', 'Face2Face', 'FaceSwap', 'NeuralTextures']
if FACES_ROOT.exists():
    for m in methods:
        d = FACES_ROOT / m
        if not d.exists():
            print('Missing method dir:', d)
            continue
        sample = next(iter(d.glob('*')), None)
        print(m, 'exists;', 'example video dir:', sample)
else:
    print('Faces root not present locally:', FACES_ROOT)
```
